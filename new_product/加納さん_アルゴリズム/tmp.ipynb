```python
# 新商品と同じカテゴリの商品一覧の中から条件にあう商品を探す
# 条件　
#   ＊最初の日が０（つまり、学習データ期間の途中から販売開始をした商品）で、
#   ＊1年１ヶ月以上（400日以上）のデータをもち、
#   ＊販売開始後、nanでも0でもない行の割合が95%以上

import pandas as pd
import numpy as np

# 条件に合う列を抽出する関数
def filter_columns_by_conditions(df):
    valid_columns = []

    for col in df.columns:
        column_data = df[col]
        
        # 最初の行がnanか0であるか確認
        if not (pd.isna(column_data.iloc[0]) or column_data.iloc[0] == 0):
            continue
        
        # 最初のnanでも0でもない行を取得
        first_valid_idx = column_data[(~column_data.isna()) & (column_data != 0)].index[0]

        # 最初の有効な行から最後までのデータを取得
        valid_data = column_data[first_valid_idx:]
        
        # 行数が400以上であることを確認
        if len(valid_data) < 400:
            continue

        # nanでも0でもない行の割合が95%以上であることを確認
        non_nan_non_zero_count = valid_data[(~valid_data.isna()) & (valid_data != 0)].count()
        if non_nan_non_zero_count / len(valid_data) < 0.95:
            continue

        # 条件を満たす場合、列名を記録
        valid_columns.append(col)

    return df[valid_columns]

# 条件に合う列を抽出
filtered_df = filter_columns_by_conditions(result)

# 結果表示
filtered_df
```
